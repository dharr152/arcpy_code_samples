# Madison Neighborhood Residential Redevelopment 
## Using proximity to parks and metro stations to identify areas of potential investment

In 2009, The city of Madison, Wisconsin adopted an extensive plan to revitalize and develop different neighborhoods within the city limits and beyond in accordance with best sustainability practices.  In accordance with this plan, specific areas were designated for specific land-uses, such as residential, commercial, agricultural, and mixed use areas[^1].   To document this process, the city maintains a database of proposed redevelopment zones and within the city of Madison.  This analysis focuses on areas designated as residential, and identifies areas of interest to residential developers based on specific characteristics the literature points to for a healthy urban residential environment.  The factors considered here are proximity to parks, and proximity to metro transit stations.  Additionally, this study seeks to identify those areas that are coded to be vacant as ideal zones to redevelop to invigorate community[^2].

**Tool Development**
The tool developed for this analysis works to merge tax assessors parcel data with the neighborhood development shapefile for the specific targeted fields of interest, and to calculate additional fields using tools from the overlay toolbox.  All necessary data for analysis is downloaded within the tool's execution and stored in a local geodatabse in a folder specified by the user.  The tool then executes a series of select operations to make sure only those areas that overlap between the parcels and neighborhood development areas that are coded as residential are included in analysis.  


The desired distances from parks and bus stops (specified by the user in the tool dialog) are then coded into 'flag fields' (containing the values of `1` or `0` depending on whether they fall within the specified distances.  An additional flag value for lots coded as `Vacant` is included as well, and finally, an "index" value is added as the sum of all of these fields.  Each field may hold a customized weight as specified by the user in the tool dialog.  The default value of each field is `1.` When this index value is symbolized in four graduated classes (suggested colors are a purple or darker green), the darkest areas are those residential plots that are within  a specified distance of parks and metro stops, and are also coded to be vacant (these are the three fields for which flags were created as specified above).   


Finally, this tool output includes in its Results.gdb an interpolated surface of property values for all fo the city of Madison's residential parcels to compare to the areas identified by this tool to be optimal for residential redevlopment.

# Using this tool

This tool is very straightforward to use, because all of the data is downloaded within the tool's execution. The central parameter for the tool is simply a file folder that is a local directory on the user's machine. 
![tool_dialog](.\img\tool_dialog.png)

The remaining parameters hold default values as shown above.  The distance to parks is specified in `Miles`, while the bus distance is specified in `feet,` because the measuring mechanism employed by this tool relies on the spatial projection, which in this case is in feet (NAD83).  The remaining three parameters, whose default value is 1, hold the relative weight of the three characteristics (distance to parks and buses, and vacancy).  As long as these values are the same, the three are weighted equally.  The relative difference between them will indicate how the index is calculated. 

There are two geodatabases created within this tool.  One is titles `Working_Archive` and is preserved for archival purposes, and the other is called `Results` and contains the two feature classes of interest for this analysis. 


It is reasonable to expect this tool to take up to one minute to execute.
